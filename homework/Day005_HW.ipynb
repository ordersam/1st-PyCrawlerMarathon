```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true')
import json
data = json.loads(r.text)
print('å›å‚³è³‡æ–™ç­†æ•¸ =', len(data), '\n')
print('è³‡æ–™æ¬„ä½ =', data[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
import numpy as np
import pandas as pd

data_columes = ['title', 'createdAt', 'commentCount', 'likeCount']
data_content = [[], [], [], []]

for d in data:
    for i in np.arange(4):
        data_content[i].append(d[data_columes[i]])

df = pd.DataFrame()
for i in np.arange(4):
    df[data_columes[i]] = data_content[i]
df
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
import json
import numpy as np
import pandas as pd
import requests

r1 = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r2 = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=false')

data1 = json.loads(r1.text)
data2 = json.loads(r2.text)

data_columes = ['commentCount', 'likeCount']
data_content1 = [[], []]
data_content2 = [[], []]

for d in data1:
    for i in range(2):
        data_content1[i].append(d[data_columes[i]])
data_mean1 = np.around(np.mean(data_content1, axis=1), 4)
print('ç†±é–€æ–‡ç« \nå¹³å‡ç•™è¨€äººæ•¸ =', data_mean1[0], '\nå¹³å‡æŒ‰è®šäººæ•¸ =', data_mean1[1], '\n')

for d in data2:
    for i in range(2):
        data_content2[i].append(d[data_columes[i]])
data_mean2 = np.around(np.mean(data_content2, axis=1), 4)
print('éç†±é–€æ–‡ç« \nå¹³å‡ç•™è¨€äººæ•¸ =', data_mean2[0], '\nå¹³å‡æŒ‰è®šäººæ•¸ =', data_mean2[1])
```
