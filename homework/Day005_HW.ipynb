## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true')
import json
data = json.loads(r.text)
print('回傳資料筆數 =', len(data), '\n')
print('資料欄位 =', data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
import numpy as np
import pandas as pd

data_columes = ['title', 'createdAt', 'commentCount', 'likeCount']
data_content = [[], [], [], []]

for d in data:
    for i in np.arange(4):
        data_content[i].append(d[data_columes[i]])

df = pd.DataFrame()
for i in np.arange(4):
    df[data_columes[i]] = data_content[i]
df
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
import json
import numpy as np
import pandas as pd
import requests

r1 = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r2 = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=false')

data1 = json.loads(r1.text)
data2 = json.loads(r2.text)

data_columes = ['commentCount', 'likeCount']
data_content1 = [[], []]
data_content2 = [[], []]

for d in data1:
    for i in range(2):
        data_content1[i].append(d[data_columes[i]])
data_mean1 = np.around(np.mean(data_content1, axis=1), 4)
print('熱門文章\n平均留言人數 =', data_mean1[0], '\n平均按讚人數 =', data_mean1[1], '\n')

for d in data2:
    for i in range(2):
        data_content2[i].append(d[data_columes[i]])
data_mean2 = np.around(np.mean(data_content2, axis=1), 4)
print('非熱門文章\n平均留言人數 =', data_mean2[0], '\n平均按讚人數 =', data_mean2[1])
```
