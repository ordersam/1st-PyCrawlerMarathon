## 作業目標

根據範例：

* 取出今天所有的新聞
* 取出今天下午三點到五點的新聞
* 根據範例，取出三天前下午三點到五點的新聞

## 根據範例，取出今天所有的新聞

```python
# 1.打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list-2019-12-23-0.htm")
    # 每兩秒自動往下滑
import time
time.sleep(10)
for i in range(100):
    time.sleep(3)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
# 2.取得資料（等瀏覽器資料出現後才執行）：丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
# 3.關閉瀏覽器
browser.quit()
# 所有新聞
time, classification, content = [], [], []
for d in soup.find(class_="part_list_2").find_all('h3'):
    import re
    match = re.findall('^(2019/12/23 )', d.find(class_="date").text)
    if match:
        time.append(d.find(class_="date").text)
        classification.append(d.find('em').text)
        content.append(d.find_all('a')[-1].text)
        print(d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text)
```

## 取出今天下午三點到五點的新聞

```python
time_3_5, classification_3_5, content_3_5 = [], [], []
for d in soup.find(class_="part_list_2").find_all('h3'):
    import re
    match = re.findall('^(2019/12/23 17:00)|(2019/12/23 16:)|(2019/12/23 15:)', d.find(class_="date").text)
    if match:
        time_3_5.append(d.find(class_="date").text)
        classification_3_5.append(d.find('em').text)
        content_3_5.append(d.find_all('a')[-1].text)
        print(d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text)
```
