### ① 取出今天所有的發文

```python
# 不使用動態網頁抓取的話，可以從各分類抓取"今天"所有文章
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
    # 所有分類名稱 + 網址
classification_name, classification_url = [], []
for d in soup.find(class_="part_menu_2").find_all('a'):
    classification_name.append(d.text)
    classification_url.append('https://www.ettoday.net'+d.get('href'))
    # 各分類分別下載資料列印
time, classification, content = [], [], []
for num, name in enumerate(classification_url):
    url_d = name
    r = requests.get(url_d)
    soup = BeautifulSoup(r.text, "html5lib")
    
    for d in soup.find(class_="part_list_2").find_all('h3'):
        import re
        match = re.findall('^(2019/12/10)', d.find(class_="date").text)
        if match:
            time.append(d.find(class_="date").text)
            classification.append(d.find('em').text)
            content.append(d.find_all('a')[-1].text)
            print(d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
df = pd.DataFrame()
df['time'] = time
df['classification'] = classification
df['content'] = content
df['count'] = 1
df
```

### ③ 哪一個類別的文章最多

```python
class_count = df.groupby('classification').size()
import numpy as np
print('今日最多文章的分類 = ' + class_count.idxmax(), '\n該分類文章篇數 = ' + str(class_count.max()))
```
