## 作業目標

利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。


### 1. https://www.zhihu.com/explore 

```python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

### 利用查看開發者工具觀察 Headers

```python
# 截圖
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
plt.figure(figsize=(24, 16))
img = mpimg.imread('0.data/zhihu_header.png')
plt.imshow(img)
```

```python
# 加上headers再次抓取，比較內容
import requests
url = 'http://www.zhihu.com'

headers = {
    'authority': 'www.zhihu.com',
    'method': 'GET',
    'path': '/api/v4/search/preset_words?w=',
    'scheme': 'https',
    'accept': '*/*',
    # 'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7,zh-CN;q=0.6',
    'referer': 'https://www.zhihu.com/explore',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-origin',
    'user-agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36',
    'x-api-version': '3.0.91',
    'x-app-za': 'OS=Web',
    'x-requested-with': 'fetch'
}

r = requests.get(url, headers=headers)
r.encoding = 'utf-8'
print(r.text[0:600])
```
