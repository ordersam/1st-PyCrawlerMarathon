## 作業目標


* 根據範例使用 selenuim 取出 高雄市小港區 2019 年 空氣品質 資料，並用一個適合的資料結構儲存
* Hint: 將所有資料存在一個變數當中



```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
# 1.執行程式打開新的瀏覽器頁面到 環保署 空氣品質監測網站
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx")
    # 選地點、年度
selectSite = Select(browser.find_element_by_id("ctl05_ddlSite"))
selectSite.select_by_value('58')     # 高雄市-小港
selectYear = Select(browser.find_element_by_id("ctl05_ddlYear"))
selectYear.select_by_value('2019')
    # 點查詢
browser.find_element_by_id('ctl05_btnQuery').click()
```

![](https://imgur.com/LwGGT8c.jpg)

```python
# 2.取得資料（等瀏覽器資料出現後才執行）
import time
time.sleep(10)

html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
soup
```

```python
table = soup.find('table', class_='TABLE_G')
table
```

```python
dict_all = {}
each_item_index_range = int(len(table.find_all('td')) / 10)
for i in range(10):
    start_index = int(each_item_index_range * i)
    end_index = int(each_item_index_range * (1+i))
    each_item = table.find_all('td')[start_index:end_index]
    item_name = each_item[0].text
    each_item_dict = {}
    for j in range(2, each_item_index_range, 3):
        each_item_dict[each_item[j].text] = each_item[j+1].text
    dict_all[item_name] = each_item_dict
dict_all
```

```python
import pandas as pd
pd.DataFrame(dict_all)
```
